USING PANDAS FOR DATA CLEANING AND REMOVING ALL NULL VALUES

```python
import pandas as pd 
movies=pd.read_csv("tmdb_5000_movies.csv")
credits=pd.read_csv("tmdb_5000_credits.csv")
movies.info()
movies.shape
```

```python
credits.info()
```

```python
movies=movies.merge(credits,on="title")
movies.info()
movies.shape
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies=movies.drop(columns="homepage")
movies=movies.drop(columns="tagline")
```

```python
movies.isnull().sum()
movies.dropna(inplace=True)
```

```python
import ast

def convert(text):
    L = []

    for i in ast.literal_eval(text):
        L.append(i['name'])

    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.dtypes
```

```python
movies['genres'].head()
```

```python
def convert_cast(text):
    L = []
    counter = 0

    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
            counter += 1
        else:
            break

    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
def fetch_director(text):
    L = []

    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break

    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])

```

```python
import ast

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
type(movies['keywords'].iloc[0])
```

```python
movies['keywords'] = movies['keywords'].apply(
    lambda x: [i.replace(" ", "") for i in x]
)
```

```python
movies['tags'] = (
    movies['overview']
    + movies['genres']
    + movies['keywords']
    + movies['cast']
    + movies['crew']
)
```

```python
movies['tags'].iloc[0]
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))
```

```python
movies['tags'].iloc[0]
```

```python
new_df=movies[['id','title','tags']]
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(
    max_features=5000,
    stop_words='english'
)
vector=cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim=cosine_similarity(vector)
print(sim)
```

```python
def recommedmovies(movie):
    movie=movie.lower()
    if movie not in movies['title'].str.lower().values:
        print("Movie not found")
        return
    index=movies[movies['title'].str.lower()==movie].index[0]
    distance=sim[index]
    movielist=sorted(
        list(enumerate(distance)),reverse=True,key=lambda x:x[1]

    )[1:6]
    for i in movielist:
        print(movies.iloc[i[0]].title)
```

```python
recommedmovies('Interstellar')
```
